```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
directory =  # path to the directory where the csv files are stored
file_names1 = ["50-5.csv", "100-10.csv", "200-20.csv", "300-30.csv", "500-50.csv", "750-75.csv", "1000-100.csv"]

data_frames1 = []
for file_name in file_names1:
    full_file_path1 = os.path.join(directory, file_name)
    df1 = pd.read_csv(full_file_path1, skiprows=1)
    mapping_value1 = df1.iloc[0, 11]
    
    filtered_data1 = df1[df1.iloc[:, 11] == mapping_value1]
    data_frames1.append(filtered_data1)
```
